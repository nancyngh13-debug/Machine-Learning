Evaluación eficiencia listas de python vs matrices numpy

```python
import sys
S= range (1000)
print("Resultado lista de python")
print(sys.getsizeof(5)*len(S))
print()

D=np.arange(1000)
print("Resultado lista de numpy")
print(D.size*D.itemsize)
```

```python
import time
SIZE =100000

L1= range(SIZE)
L2= range(SIZE)
A1=np.arange(SIZE)
A2=np.arange(SIZE)

start=time.time()
Result=[(x,y) for x,y in zip(L1,L2)]
print("Resultado Lista python:")
print((time.time()-start)*1000)
print()

start=time.time()
Result=A1+A2
print("Resultado Numpy array:")
print((time.time()-start)*1000)
```

```python
#Crear una matriz de unas -3 filas y 4 columnas
unos = np.ones((3,4))
print(unos)
```

```python
#Crear una matriz de ceros - 3 filas 4 columnas
ceros = np.zeros((3,4))
print(ceros)
```

```python
#Crear una matriz con valores aleatorios
aleatorios =np.random.random ((2,2))
print(aleatorios)
```

```python
#Crear matriz vacia
vacia = np.empty((3,2))
print(vacia)
```

```python
#Crear una matriz con un solo valor
full = np.full((2,2),8)
print(full)
```

```python
#crear una matriz con valores espaciados uniformemente
espacio1 = np.arange(0,30,5)
print(espacio1)
espacio2 = np.linspace(0,5,10)
print(espacio2)
```

```python
#Crear una matriz identidad
identidad1 = np.eye(4,4)
print(identidad1)
identidad2 = np.identity(4)
print(identidad2)
#
```

inspeccionar matrices

```python
#Conocer las dimenciones de una matriz
a=np.array([(1,2,3),(4,5,6)])
print(a.ndim)
```

```python
#Consular el tipo de datos
a=np.array([(1,2,3)])
print(a.dtype)
```

```python
#Consultar el tamaño y la forma de la matriz
a=np.array([(1,2,3,4,5,6)])
print(a.size)
print(a.shape)
```

```python
#Cambio de forma de una matriz
a=np.array([(8,9,10),(11,12,13)])
print(a)
a=a.reshape(3,2)
print(a)
#
```

```python
#Extraer un solo valor de la matriz - el valor ubicado en la fina 0 comumna 2
a = np.array([(1,2,3,4),(3,4,5,6)])
print(a[0,2])
```

```python
#Extraeer los valores de todas las filas ubicadas en la columna 3
a=np.array([(1,2,3,4),(3,4,5,6)])
print(a[0:,2])
```
